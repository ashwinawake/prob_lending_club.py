```python
"""Write a script called "prob_lending_club.py" that reads in the loan data, cleans it, and loads it into a pandas DataFrame. 
Use the script to generate and save a boxplot, histogram, and QQ-plot for the values in the "Amount.Requested" column. 
Be able to describe the result and how it compares with the values from the "Amount.Funded.By.Investors".
Push your code to GitHub and enter the link below."""

import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

loansData = pd.read_csv('https://github.com/Thinkful-Ed/curric-data-001-data-sets/raw/master/loans/loansData.csv')
print(loansData)
```

```python
#cleaning data, removing null values
loansData.dropna(inplace=True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,10))
loansData.boxplot(column = 'Amount.Requested', ax=axes[0])
loansData.boxplot(column = 'Amount.Funded.By.Investors', ax=axes[1])

"""Inferences:
By observing the box plots of Amount.Requested and Amount.Funded.By.Investors we can infer the following:

1. The Highest value in Amount.Requested is closer to 35,000 and the highest value in Amount.Funded.By.Investors is a little 
above 30,000

2. The data is a little more spread out in Amount.Requested than Amount.Funded.By.Investors(Not by a wide margin).

3. The median for both data sets is at 10,000.

4. The interquartile range for Amount.Requested appears to be slightly larger than Amount.Funded.By.Investors.
   i.e A good chunk of the data lie between Q2 and Q3 for both data sets.They aren't centered.

5. The median values for quartiles 1 and 2 in the box plot are a little over 5,000 for both data sets and for quartiles 3 and 4 
   they are at 17,500 for Amount.Requested and 16,000 for Amount.Funded.By.Investors.
   
6. Clearly there is a gap between the Amount Requested and Amount being funded by an investor.
"""

```
